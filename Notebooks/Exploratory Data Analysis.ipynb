###  Data Description and Summary Requirement

#### **1. purchases Table**
- Contains actual purchase data.
- Includes:
  - **Date of purchase**
  - **Products (brands)** purchased by vendors
  - **Amount paid (in dollars)**
  - **Quantity purchased**

#### **2. purchase_prices Table**
- Provides product-wise **actual** and **purchase prices**.
- The **combination of vendor and brand** is unique in this table.
- The `PurchasePrice` column in the `purchases` table is **derived** from this table.

#### **3. vendor_invoice Table**
- Aggregates data from the `purchases` table.
- Summarizes:
  - **Quantity**
  - **Dollar amounts**
  - **Freight costs**
- Maintains uniqueness based on **vendor** and **PO number**.

#### **4. sales Table**
- Captures actual sales transactions.
- Includes:
  - **Brands** purchased by vendors
  - **Quantity sold**
  - **Selling price**
  - **Revenue earned**

###  Objective
Since the required data for analysis is distributed across multiple tables, we need to **create a summary table** that consolidates:

- Purchase transactions made by vendors  
- Sales transaction data  
- Freight costs for each vendor  
- Actual product prices from vendors


```python
frieght_summmary = pd.read_sql('''select VendorNumber,sum(Freight)as freightcost from 
vendor_invoice group by VendorNumber''',engine)
frieght_summmary
```

```python
purchase_summary = pd.read_sql("""SELECT 
    p.VendorNumber,
    p.VendorName,
    p.Brand,
    p.Description,
    p.PurchasePrice,
    pp.Price as ActualPrice,
    pp.Volume,
    SUM(p.Quantity) AS TotalPurchaseQuantity,
    SUM(p.Dollars) AS TotalPurchaseDollars
FROM purchases p
JOIN purchase_prices pp
    ON p.Brand = pp.Brand
WHERE p.PurchasePrice > 0
GROUP BY p.VendorNumber,p.VendorName, p.Brand, p.Description, p.PurchasePrice, pp.Price, pp.Volume
ORDER BY TotalPurchaseDollars
""",engine)
purchase_summary
```

```python
sales_summary = pd.read_sql_query("""SELECT 
    VendorNo,
    Brand,
    SUM(SalesQuantity) AS TotalSalesQuantity,
    SUM(SalesDollars) AS TotalSalesDollars,
    SUM(SalesPrice) AS TotalSalesPrice,
    SUM(ExciseTax) AS TotalExciseTax
FROM sales
GROUP BY VendorNo, Brand""",engine)
sales_summary
```

```python
# merging all the summary table to create an aggregated summary table for further analysis
vendor_sales_summary = pd.read_sql_query("""WITH FreightSummary AS (
    SELECT 
        VendorNumber, 
        SUM(Freight) AS FreightCost 
    FROM vendor_invoice 
    GROUP BY VendorNumber
), 

PurchaseSummary AS (
    SELECT 
        p.VendorNumber,
        p.VendorName,
        p.Brand,
        p.Description,
        p.PurchasePrice,
        pp.Price AS ActualPrice,
        pp.Volume,
        SUM(p.Quantity) AS TotalPurchaseQuantity,
        SUM(p.Dollars) AS TotalPurchaseDollars
    FROM purchases p
    JOIN purchase_prices pp
        ON p.Brand = pp.Brand
    WHERE p.PurchasePrice > 0
    GROUP BY p.VendorNumber, p.VendorName, p.Brand, p.Description, p.PurchasePrice, pp.Price, pp.Volume
), 

SalesSummary AS (
    SELECT 
        VendorNo,
        Brand,
        SUM(SalesQuantity) AS TotalSalesQuantity,
        SUM(SalesDollars) AS TotalSalesDollars,
        SUM(SalesPrice) AS TotalSalesPrice,
        SUM(ExciseTax) AS TotalExciseTax
    FROM sales
    GROUP BY VendorNo, Brand
) 

SELECT 
    ps.VendorNumber,
    ps.VendorName,
    ps.Brand,
    ps.Description,
    ps.PurchasePrice,
    ps.ActualPrice,
    ps.Volume,
    ps.TotalPurchaseQuantity,
    ps.TotalPurchaseDollars,
    ss.TotalSalesQuantity,
    ss.TotalSalesDollars,
    ss.TotalSalesPrice,
    ss.TotalExciseTax,
    fs.FreightCost
FROM PurchaseSummary ps
LEFT JOIN SalesSummary ss 
    ON ps.VendorNumber = ss.VendorNo 
    AND ps.Brand = ss.Brand
LEFT JOIN FreightSummary fs 
    ON ps.VendorNumber = fs.VendorNumber
ORDER BY ps.TotalPurchaseDollars DESC""",engine)
```

###  Vendor-wise Sales and Purchase Summary

This query generates a **vendor-wise sales and purchase summary**, which is valuable for the following reasons:

---

####  Performance Optimization

- The query involves **heavy joins and aggregations** on large datasets like *sales* and *purchases*.  
- **Storing pre-aggregated results** helps avoid repeated expensive computations.  
- It assists in analyzing **sales, purchases, and pricing trends** for different vendors and brands.  

---

####  Future Benefits for Dashboarding & Reporting

- Instead of running complex queries each time, dashboards can **fetch data quickly** from the `vendor_sales_summary` table.  
- This enables **faster insights**, improved **report performance**, and **reduced database load**.  

---

####  Data Cleaning

Before storing the summary, we should **clean the data** to ensure there are no inconsistencies or missing values.
This step improves data quality and ensures accurate reporting results.

```python
import numpy as np
vendor_sales_summary.replace([np.inf, -np.inf], np.nan, inplace=True)
vendor_sales_summary.fillna(0,inplace = True)
```

```python
vendor_sales_summary['VendorNumber'] = vendor_sales_summary['VendorNumber'].astype(int)   
vendor_sales_summary['Brand'] = vendor_sales_summary['Brand'].astype(int)      
vendor_sales_summary['PurchasePrice'] = vendor_sales_summary['PurchasePrice'].astype(float)    
vendor_sales_summary['ActualPrice'] = vendor_sales_summary['ActualPrice'].astype(float)       
vendor_sales_summary['Volume'] = vendor_sales_summary['Volume'].astype(float)       
vendor_sales_summary['TotalSalesQuantity'] = vendor_sales_summary['TotalSalesQuantity'].astype(int)  
```

```python
vendor_sales_summary['VendorName'] = vendor_sales_summary['VendorName'].str.strip()
```

```python
# creating new columns for better analysis
vendor_sales_summary['GrossProfit'] = vendor_sales_summary['TotalSalesDollars'] - vendor_sales_summary['TotalPurchaseDollars']
vendor_sales_summary['ProfitMargin'] = (vendor_sales_summary['GrossProfit'] / vendor_sales_summary['TotalSalesDollars'])*100
vendor_sales_summary['StockTurnover'] = vendor_sales_summary['TotalSalesQuantity'] / vendor_sales_summary['TotalPurchaseQuantity']
vendor_sales_summary['SalesToPurchaseRatio'] = vendor_sales_summary['TotalSalesDollars'] / vendor_sales_summary['TotalPurchaseDollars']
```
